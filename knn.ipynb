```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder


df = pd.read_csv('/content/iphone_purchase_records.csv')


label_encoder = LabelEncoder()

df['Gender_Encoded'] = label_encoder.fit_transform(df['Gender'])


print(df.head())
```

```python
df = df.drop('Gender', axis=1)
```

```python
print(df.head())
```

```python
X = df[['Age', 'Salary', 'Gender_Encoded']]  # Select features for X
Y = df['Purchase Iphone']  # Select target for Y

print(X.head())  # To see the first few rows of X
print(Y.head())  # To see the first few rows of Y
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y,test_size=0.2,random_state=7)
```

```python
clf = KNeighborsClassifier()
clf.fit(X_train,Y_train)
```

```python
score = clf.score(X_train,Y_train)
print(clf.predict([[19,19000,1]]))
```

```python
predict = clf.predict(X_test)
print(classification_report(Y_test, predict))
```
